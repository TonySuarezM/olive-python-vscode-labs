# M0 — Setup Check (Olive + ONNX Runtime)

Este notebook verifica:

- Qué Python está ejecutando el kernel
- Versiones de paquetes (`olive-ai`, `onnxruntime`, `onnx`, `numpy`)
- `pip check` (dependencias rotas)
- Imports básicos (`olive`, `onnxruntime`)


```python
import sys
print("Python executable:", sys.executable)
print("Python version:", sys.version)
```

```python
import subprocess, sys

def pip_show(pkg: str):
    print(f"\n--- pip show {pkg} ---")
    subprocess.run([sys.executable, "-m", "pip", "show", pkg], check=False)

for pkg in ["olive-ai", "onnxruntime", "onnx", "numpy", "ipykernel"]:
    pip_show(pkg)
```

```python
import subprocess, sys
print("\n--- pip check ---")
subprocess.run([sys.executable, "-m", "pip", "check"], check=False)
```

```python
import olive
import onnxruntime as ort

print("Olive import OK:", olive.__name__)
print("ONNX Runtime version:", ort.__version__)
print("Available providers:", ort.get_available_providers())
```
